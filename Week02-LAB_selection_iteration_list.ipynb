5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python

marks = [65, 78, 45, 90, 55, 30, 85]
total = 0
for m in marks:
    total += m
average = total / len(marks)

highest = max(marks)
lowest = min(marks)

passed = sum(1 for m in marks if m >= 50)
failed = sum(1 for m in marks if m < 50)

print("Average mark:", average)
print("Highest mark:", highest)
print("Lowest mark:", lowest)
print("Passed:", passed)
print("Failed:", failed)
````

Write a program that implements the conversion from human years to dog years described in the previous paragraph. 16. Write a **guessing game** in Python where the user has to guess a secret number. After every guess the program tells the user whether their number was too large or too small. At the end the number of tries needed should be printed.


````python
import random
secret_number = random.randint(1, 100)
tries = 0

while True:
    guess = int(input("Guess the number (1–100): "))
    tries += 1

    if guess < secret_number:
        print("Too small!")
    elif guess > secret_number:
        print("Too large!")
    else:
        print(f"Correct! You guessed it in {tries} tries.")
        break
````

17. **Collecting Positive Numbers**
You are writing a small program to help a teacher keep track of positive numbers entered by students. The teacher also wants to know if any even numbers were given.

**Requirements:**

1. Keep asking the user to enter a positive number, or type `"done"` to stop.
2. Store all entered numbers in a **list**.
3. After the input ends:

   * If numbers were entered:

     * Print all the numbers entered.
     * Check the list and print the **first even number**. If no even number exists, print `"No even numbers found"`.
   * If no numbers were entered, print `"No numbers were entered"`.


> ### Note 
>
>* **`lower()`**: Converts any text into lowercase.
>  Example: `"DONE".lower()` becomes `"done"`.
>  This makes it easier to compare user input without worrying about uppercase/lowercase.
>
>* **`isdigit()`**: Checks if a string only contains digits (0–9).
>  Example: `"123".isdigit()` returns `True`, but `"12a".isdigit()` returns `False`.
>  This helps us make sure the user typed a valid number before converting it into an integer.
> 

---

````python
numbers = []

while True:
    user_input = input("Enter a positive number or type 'done' to stop: ").lower()

    if user_input == "done":
        break
    elif user_input.isdigit():
        num = int(user_input)
        if num > 0:
            numbers.append(num)
        else:
            print("Please enter a positive number.")
    else:
        print("Invalid input. Please enter a number or 'done'.")

if numbers:
    print("Numbers entered:", numbers)
    even_found = False
    for n in numbers:
        if n % 2 == 0:
            print("First even number:", n)
            even_found = True
            break
    if not even_found:
        print("No even numbers found.")
else:
    print("No numbers were entered.")
````

18. Suppose you want to develop a **program to play a lottery.**
    
The program randomly generates a two-digit number, prompts the user to enter a <u>two-digit number</u>, and determines whether the user wins according to the following rules:


    1. If the user’s input matches the lottery in the exact order, the award is £10,000.
    2. If all the digits in the user’s input match all the digits in the lottery number, the award is £3,000.
    3. If one digit in the user’s input matches a digit in the lottery number, the award is £1,000.



````python
import random
lottery = random.randint(10, 99)
guess = int(input("Enter your 2-digit number: "))
lottery_str = str(lottery)
guess_str = str(guess)
print("Lottery number:", lottery)
if guess_str == lottery_str:
    print("Exact match! You win £10,000!")
elif sorted(guess_str) == sorted(lottery_str):
    print("All digits match! You win £3,000!")
elif any(digit in lottery_str for digit in guess_str):
    print("One digit matches! You win £1,000!")
else:
    print("No match. Better luck next time!")
````


21. **Password cracker** - write a program to print the accuracy of password guess.

For example: The actual password is `MDX123` while a user tried `MDX321` or `mdx123`, the program should tell that how accurately the password was guessed, in this case it should be `50%` accuracy. if a user enter `MDX124`, the output should be `83%` and in case of `123MDX` the output should be `0%`.

Points to keep in mind while calculating the accuracy a) case of the letters entered b) order of the letters

If a password is accurately guessed, the output should state `Password cracked!`

````python
actual = input("Enter the actual password: ")
guess = input("Enter the guessed password: ")
if guess == actual:
    print("Password cracked!")
else:
    matches = 0
    length = min(len(actual), len(guess))
    for i in range(length):
        if actual[i] == guess[i]:
            matches += 1
    accuracy = (matches / len(actual)) * 100
    print(f"Password accuracy: {accuracy:.0f}%")
````
