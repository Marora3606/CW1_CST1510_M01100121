### Activity 3: Temperature Logger

Write a program that:
1. Asks the user for temperatures for 5 days
2. Writes each temperature to "temps.txt" with the day number
   Format: "Day 1: 22°C"
3. Reads the file and calculates the average temperature

<hr>

`````python
num_days = 5
temperatures = []

# 1. Asks the user for temperatures for 5 days
print("Please enter temperatures for 5 days (e.g., 22.5):")
for i in range(1, num_days + 1):
    while True:
        try:
            temp_str = input(f"Day {i} temperature (°C): ")
            temp = float(temp_str)
            temperatures.append(temp)
            break
        except ValueError:
            print("Invalid input. Please enter a number.")

# 2. Writes each temperature to "temps.txt"
with open("temps.txt", "w") as file:
    for i, temp in enumerate(temperatures):
        file.write(f"Day {i+1}: {temp}°C\n")

print("--- Temperature Analysis ---")
total_temp = 0
read_temps_count = 0

# 3. Reads the file and calculates the average temperature
with open("temps.txt", "r") as file:
    for line in file:
        try:
            temp_str = line.split(": ")[1].split("°C")[0]
            temp = float(temp_str)
            total_temp += temp
            read_temps_count += 1
        except IndexError:
            continue

if read_temps_count > 0:
    average_temp = total_temp / read_temps_count
    print(f"Total temperature recorded: {total_temp:.1f}°C")
    print(f"Number of days recorded: {read_temps_count}")
    print(f"Average temperature: {average_temp:.1f}°C")
else:
    print("No valid temperatures were found in the file.")
`````

**Steps:**
1. Read the file and store each item in a set.
2. Keep track of duplicate entries.
3. Print any duplicates found.
<hr>

`````python
# Simulated content for clothing_stock.txt
# Includes duplicates: T-shirt,S and Socks,XS
clothing_stock_content_dupes = (
    "T-shirt,S\n"
    "Jeans,M\n"
    "Jacket,L\n"
    "Socks,XS\n"
    "Sweater,XL\n"
    "T-shirt,S\n"  # Duplicate
    "Socks,XS\n"   # Duplicate
    "Hat,M\n"
)
with open("clothing_stock.txt", "w") as f:
    f.write(clothing_stock_content_dupes)

seen_items = set()
duplicates = set()

print("--- Identifying Duplicate Clothing Stock ---")

# 1. Read the file and store each item in a set.
# 2. Keep track of duplicate entries.
try:
    with open("clothing_stock.txt", "r") as file:
        for line in file:
            item_entry = line.strip()
            if item_entry:
                if item_entry in seen_items:
                    duplicates.add(item_entry)
                else:
                    seen_items.add(item_entry)

    # 3. Print any duplicates found.
    if duplicates:
        print("Duplicate entries found:")
        for dupe in duplicates:
            print(f"- {dupe}")
    else:
        print("No duplicate entries found in the clothing stock.")
        
except FileNotFoundError:
    print("Error: The file 'clothing_stock.txt' was not found.")
`````

### Activity 14: Counting Different Sizes per Clothing Item

Create a Python program that reads data from the `clothing_stock.txt` file and counts how many different sizes of each clothing item are available. Store this in a dictionary and print it.

**Steps:**
1. Read the file and extract both clothing items and sizes.
2. Count how many different sizes exist for each item.
3. Store the results in a dictionary and print it.
<hr>

`````python
# Simulated content for clothing_stock.txt
clothing_stock_content = (
    "T-shirt,S\n"
    "T-shirt,M\n"
    "T-shirt,L\n"
    "Jeans,M\n"
    "Jeans,L\n"
    "Jeans,L\n" # Duplicate size for Jeans (will be ignored by set)
    "Jacket,XL\n"
    "Socks,XS\n"
    "Socks,XS\n"
)
with open("clothing_stock.txt", "w") as f:
    f.write(clothing_stock_content)

# Dictionary to store {item: set_of_unique_sizes}
item_sizes = {}

# 1. Read the file and extract both clothing items and sizes.
try:
    with open("clothing_stock.txt", "r") as file:
        for line in file:
            line = line.strip()
            if line:
                try:
                    item, size = line.split(',')
                    
                    # Initialize a set for the item if it's new
                    if item not in item_sizes:
                        item_sizes[item] = set()
                        
                    # Add the size to the set (sets automatically handle uniqueness)
                    item_sizes[item].add(size)
                except ValueError:
                    print(f"Skipping malformed line: {line}")
                    
    print("--- Unique Sizes per Clothing Item ---")
    print("Item: Count of Unique Sizes")
    
    # 2. Count how many different sizes exist for each item.
    # 3. Store the results in a dictionary and print it.
    unique_size_counts = {}
    for item, sizes_set in item_sizes.items():
        count = len(sizes_set)
        unique_size_counts[item] = count
        print(f"{item}: {count}")

except FileNotFoundError:
    print("Error: The file 'clothing_stock.txt' was not found.")
`````

### Activity 15: Managing Study Groups

You are responsible for managing a study group for a course. Each student can be part of multiple study groups, and you need to track which groups a student is enrolled in. Students can join or leave study groups. What data structure would you choose to manage this efficiently, and why?

**Example data:**

```python
{"John": ["Math", "Physics"], "Emily": ["Physics", "Biology"], "Alex": ["Math"]}

Steps:

1. Choose a data structure that allows quick access to the groups each student belongs to.
2. Write a program that lets students join and leave study groups.
3. Implement a function to display all groups a student is part of.
<hr>

`````python
# 1. Choose a data structure (Dictionary with List values)
study_groups = {
    "John": ["Math", "Physics"], 
    "Emily": ["Physics", "Biology"], 
    "Alex": ["Math"]
}

def display_groups(groups, student):
    if student in groups:
        if groups[student]:
            print(f"{student} is in groups: {', '.join(groups[student])}")
        else:
            print(f"{student} is not currently in any study group.")
    else:
        print(f"Student '{student}' not found.")
        
def join_group(groups, student, group_name):
    groups.setdefault(student, []).append(group_name)
    print(f"\n{student} joined {group_name}.")

def leave_group(groups, student, group_name):
    if student in groups and group_name in groups[student]:
        groups[student].remove(group_name)
        print(f"\n{student} left {group_name}.")
    else:
        print(f"\nError: {student} is not in {group_name}.")

print("--- Study Group Manager ---")
print("Initial Groups:")
display_groups(study_groups, "John")

# 2. Write a program that lets students join and leave study groups.
join_group(study_groups, "Alex", "Physics")
leave_group(study_groups, "Emily", "Biology")

# 3. Implement a function to display all groups a student is part of.
print("\n--- Current Groups Status ---")
display_groups(study_groups, "Alex")
display_groups(study_groups, "Emily")
display_groups(study_groups, "John")
`````

Read the content from `Les_Brown.txt`, and then count the number of occurrences of each word in the text. Write the word count results to a new file called `word_count.txt`.

**Steps:**
1. Open and read the contents of `Les_Brown.txt`.
2. Split the content into words and count the frequency of each word.
3. Write the word count in the format: `word: count` to a new file called `word_count.txt`.

<hr>

`````python
# Week 5 – Word Count Program (Edited for Workshop Level)

# 1. Try to open and read the contents of Les_Brown.txt
try:
    with open("Les_Brown.txt", "r") as file:
        content = file.read().lower()  # read and convert text to lowercase
except FileNotFoundError:
    print("Error: File not found. Please make sure 'Les_Brown.txt' exists.")
    content = ""

# 2. Remove basic punctuation (simple cleaning)
punctuation_marks = [",", ".", "!", "?", ":", ";", "'", '"']
for mark in punctuation_marks:
    content = content.replace(mark, "")

# 3. Split text into individual words
words = content.split()

# 4. Count how many times each word appears
word_counts = {}
for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# 5. Write the word count results to a new file
try:
    with open("word_count.txt", "w") as outfile:
        for word in sorted(word_counts.keys()):
            outfile.write(f"{word}: {word_counts[word]}\n")
    print("\nWord counts have been written to 'word_count.txt'.")
except:
    print("Error: Could not write to 'word_count.txt'.")

# 6. Display summary results
print(f"\nTotal unique words: {len(word_counts)}")

# Find and display the top 3 most frequent words (simplified method)
sorted_words = sorted(word_counts.items(), key=lambda item: item[1], reverse=True)

print("\nTop 3 most frequent words:")
for word, count in sorted_words[:3]:
    print(f"- {word}: {count}")
`````
